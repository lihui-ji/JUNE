```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style
import matplotlib as mpl
mpl.style.use('default')
```

```python
def extractXcenter (df):
    result = np.zeros([10,10])
    for indx in range(10):
        for indy in range(10):
            for indz in range(10):
                if (indx == 5):
                    ind = 100 * indz + 10 * indy + indx
                    result[indz][indy] = df[ind]
    return result

def extractYcenter (df):
    result = np.zeros([10,10])
    for indx in range(10):
        for indy in range(10):
            for indz in range(10):
                if (indy == 5):
                    ind = 100 * indz + 10 * indy + indx
                    result[indz][indx] = df[ind]
    return result

def extractZcenter (df):
    result = np.zeros([10,10])
    for indx in range(10):
        for indy in range(10):
            for indz in range(10):
                if (indz == 5):
                    ind = 100 * indz + 10 * indy + indx
                    result[indy][indx] = df[ind]
    return result

def readData (lines):
    width = len(lines[0].replace('\n','').split(' '))
    length = len (lines)
    df = pd.DataFrame(np.zeros([length,width]))
    for ind in range(length):
        curr = lines[ind].replace('\n','').split(' ')
        for dimind in range(width):
            df.iloc[ind,dimind] = float(curr[dimind])
    return df
```

```python
with open(r"\\wsl.localhost\Ubuntu\home\lihuiji\JUNE\solvers\make\log.cavity.steady") as f:
    lines = f.readlines()
Uresults = []
Presults = []
Ustarts = []
Pstarts = []
for ind in range(len(lines)):
    if lines[ind] == 'CellField 1000 X 3\n':
        Ustarts.append(ind+1)
for start in Ustarts:
    Uresults.append(readData(lines[start:start+1000]))

for ind in range(len(lines)):
    if lines[ind] == 'CellField 1000 X 1\n':
        Pstarts.append(ind+1)
for start in Pstarts:
    Presults.append(readData(lines[start:start+1000]))
```

```python
def round_sig(x, sig=1):
    if x == 0:
        return 0
    return round(x, sig-int(np.floor(np.log10(abs(x))))-1)
```

```python
ary = np.array(Uresults[-1].iloc[:,0])
fig, (ax1, ax2, ax3, cax) = plt.subplots(ncols=4,figsize=(10,3), 
                  gridspec_kw={"width_ratios":[2,2,2, 0.1]})
minmin = round(np.min(ary),0)
maxmax = round(np.max(ary),0)
im = ax1.imshow(extractXcenter(ary)[::-1], vmin = minmin, vmax = maxmax, cmap='hot', interpolation='bilinear')
ax1.set_title("X crosssection")
ax2.imshow(extractYcenter(ary)[::-1], vmin = minmin, vmax = maxmax, cmap='hot', interpolation='bilinear')
ax2.set_title("Y crosssection")
ax3.imshow(extractZcenter(ary)[::-1], vmin = minmin, vmax = maxmax, cmap='hot', interpolation='bilinear')
ax3.set_title("Z crosssection")
fig.colorbar(im, cax=cax)
plt.savefig(saveFolder + r"\Uxsteady.png")
```
